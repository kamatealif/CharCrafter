```python
# creating a bigram of words 
b ={}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch, ch1 in zip(chs, chs[1:]):
        biagram = (ch, ch1)
        b[biagram] = b.get(biagram, 0) + 1
```

```python
# create a tensor of the string names 
import torch
```

```python
N = torch.zeros(size=(28,28),dtype=torch.int32)
N
```

```python
chars = sorted(list(set(''.join(words))))


string_to_int = {s:i for i,s in enumerate(chars)}
string_to_int['<S>'] =26
string_to_int['<E>'] =27
string_to_int
```

```python
# create a tensor of the string names
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        iX1 = string_to_int[ch1]
        iX2 = string_to_int[ch2]
        N[iX1, iX2] += 1;
```

```python
N
```

```python
# lets see our data into the plots
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plot looks to ugly so lets make it better
# inver the string_to_int
int_to_string = {i:s for s,i in string_to_int.items()}
int_to_string
```

```python
plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(28):
    for j in range(28):
        chstr = int_to_string[i] + int_to_string[j]
        plt.text(j, i, chstr, ha='center', va='center', fontsize=8, color='gray')
        plt.text(j,i, N[i,j].item(), ha='center', va='center', fontsize=8, color='gray')
plt.axis('off')
```
