```python
# creating a bigram of words 
b ={}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch, ch1 in zip(chs, chs[1:]):
        biagram = (ch, ch1)
        b[biagram] = b.get(biagram, 0) + 1
```

```python
sorted(b.items(), key=lambda kv: kv[1])

# if you in desc order 
sorted(b.items(), key=lambda kv: kv[1], reverse=True)
```

```python
# create a tensor of the string names 
import torch
```

```python
N = torch.zeros(size=(28,28),dtype=torch.int32)
N
```

```python
chars = sorted(list(set(''.join(words))))


string_to_int = {s:i for i,s in enumerate(chars)}
string_to_int['<S>'] =26
string_to_int['<E>'] =27
string_to_int
```

```python
# create a tensor of the string names
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        iX1 = string_to_int[ch1]
        iX2 = string_to_int[ch2]
        N[iX1, iX2] += 1;
```
